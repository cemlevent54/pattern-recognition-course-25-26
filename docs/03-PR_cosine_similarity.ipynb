```python
x=np.array([2,1])
y=np.array([-2,5])

norm_x = np.sqrt( np.dot(x.T, x))
norm_y = np.sqrt( np.dot(y.T, y))

dot_x_y =  np.dot(x.T, y)
#np.abs(np.dot(x.T, y)) = norm_x*norm_y*cosQ

import math

angle_between_x_y = math.acos(dot_x_y / (norm_x *norm_y))
print('angle_between_x_y:', angle_between_x_y)
cos_q = math.cos(angle_between_x_y)
print('cosQ (similarity)', cos_q)
```
